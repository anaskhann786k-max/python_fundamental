```python
##### write a program to tke user input to store an employe record data returning thee  record as they are enter by the user

def employee_record():
    # Store records in a list
    records = []
    
    while True:
        print("\n--- Enter Employee Details ---")
        emp_id = input("Enter Employee ID (or type 'exit' to stop): ")
        if emp_id.lower() == 'exit':
            break
            
        name = input("Enter Name: ")
        dept = input("Enter Department: ")
        salary = input("Enter Salary: ")
        
        current_record = {
            "ID": emp_id,
            "Name": name,
            "Department": dept,
            "Salary": salary
        }
        
        records.append(current_record)
        
        print(f"\nRecord Added: {current_record}")

    print("\nFinal Employee Records List:")
    for rec in records:
        print(rec)

if __name__ == "__main__":
    employee_record()

```

```python
# write a program creat a quiz game implimwentation while loop given the user three chance to  chose and  incorrrect option and when the fourth incorrect is chosen by the user the game end 
# Quiz game with 3 lives (4th incorrect answer ends game)
questions = [
    {"q": "What is the capital of France?", "a": "Paris"},
    {"q": "What is 5 + 7?", "a": "12"},
    {"q": "What planet is known as the Red Planet?", "a": "Mars"},
    {"q": "Which ocean is the largest?", "a": "Pacific"}
]

incorrect_count = 0
question_index = 0
score = 0

# Game continues while incorrect answers < 4 and questions remain
while incorrect_count < 4 and question_index < len(questions):
    print(f"\nQuestion: {questions[question_index]['q']}")
    user_answer = input("Your answer: ").strip()

    if user_answer.lower() == questions[question_index]['a'].lower():
        print("Correct!")
        score += 1
        question_index += 1
    else:
        incorrect_count += 1
        chances_left = 4 - incorrect_count
        if incorrect_count < 4:
            print(f"Incorrect. You have {chances_left} chances left.")
        else:
            print("Incorrect. No chances left!")

# Final game outcome
if incorrect_count >= 4:
    print("\nGAME OVER: You reached the maximum number of incorrect answers.")
else:
    print("\nCongratulations! You finished the quiz.")

print(f"Final Score: {score}/{len(questions)}")
```

```python
# write a program implement a program combining and loop and condition statement together with nested loop and condition statement
# Outer loop iterating from 1 to 10
for num in range(1, 11):
    # Outer condition to check if the number is even or odd
    if num % 2 == 0:
        status = "Even"
    else:
        status = "Odd"
    
    # Nested condition: Only check for primality if number is greater than 1
    is_prime = True
    if num > 1:
        # Nested loop to check factors for primality
        for i in range(2, int(num**0.5) + 1):
            # Nested condition inside the nested loop
            if num % i == 0:
                is_prime = False
                break
    else:
        is_prime = False  # 1 is not a prime number

    # Output the combined findings
    prime_status = "and Prime" if is_prime else "and Not Prime"
    print(f"Number {num} is {status} {prime_status}")

```

```python
#creat a dictionary of employ record having key have name age gender year of experince position and domain and salary find out the average salary , maximum salary of each domain 
employees = [
    {"name": "Aman", "age": 28, "gender": "Male", "experience": 5, "position": "Developer", "domain": "IT", "salary": 60000},
    {"name": "Sara", "age": 32, "gender": "Female", "experience": 7, "position": "Manager", "domain": "HR", "salary": 80000},
    {"name": "John", "age": 26, "gender": "Male", "experience": 3, "position": "Analyst", "domain": "Finance", "salary": 50000},
    {"name": "Neha", "age": 30, "gender": "Female", "experience": 6, "position": "Developer", "domain": "IT", "salary": 75000},
    {"name": "Raj", "age": 35, "gender": "Male", "experience": 10, "position": "Manager", "domain": "Finance", "salary": 90000}
]
total_salary = 0

for emp in employees:
    total_salary += emp["salary"]

avg_salary = total_salary / len(employees)
print("Average Salary:", avg_salary)

max_salary_domain = {}

for emp in employees:
    domain = emp["domain"]
    salary = emp["salary"]
    
    if domain not in max_salary_domain:
        max_salary_domain[domain] = salary
    else:
        if salary > max_salary_domain[domain]:
            max_salary_domain[domain] = salary

print("Maximum Salary by Domain:", max_salary_domain)

```
